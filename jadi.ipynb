```python
class people():

    def __init__(self,name,code_meli):
        self.name=name
        self.code_meli=code_meli
    def info(self):
        a=f"{self.name} certificate number is {self.code_meli}"
        print(a)
```

```python
people1= people("ali karimian",1295823)
people1.info()
```

```python
class students(people):

    def __init__(self,name,code_meli,maghta,kelas):
        self.maghta=maghta
        self.kelas=kelas
        self.name=name
        self.code_meli=code_meli
    
    def student_info(self):
        s=f"{self.name} certificate number is {self.code_meli} and is {self.maghta} level {self.kelas}"
        print(s)
```

```python
s1=students("maryam mohamadi","037855285","highschool",2)
s1.student_info()
```
